One way to estimate $\pi$ is to imagine a quarter circle of radius 1 centered at the origin,  
and then throwing darts randomly at points withing the bounding square. Here's a plot of the  
unit square and a unit radius quarter circle; then I throw 200 darts at the square and  
plot the hits as green dots and the misses (outside the circle) as red dots. 

```python
circle1=plt.Circle((0,0),1,color='g', ec = 'b', linewidth=3, alpha =0.1)
fig = plt.figure()
ax = fig.add_subplot(111, aspect = 'equal')
ax.add_artist(circle1)
plt.xlim(0,1)
plt.ylim(0,1)
#
# Now plot some points to illustrate the idea:
#
N = 600
x = np.random.ranf([N])
y = np.random.ranf([N])
r = x**2 + y**2

for index, value in np.ndenumerate(r):
    if value <= 1.0:
        ax.scatter(x[index], y[index], marker = 'o', color = 'g')
    else:
        ax.scatter(x[index], y[index], marker = 'o', color = 'r')

plt.xlabel(r"x (arb. units)", size=24)
plt.ylabel(r"y (arb. units)", size = 24)
plt.show()
#fig.savefig('throwDarts.pdf', dpi = 600)
```

Now, we expect that in the limit where the number of darts, $N\rightarrow \infty$, that  
the ratio of the number of hits to the total number of darts thrown should be the same  
as the ratio of the areas. The ratio of the areas (theoretically speaking) is, of course  
$ \pi/4$, and hence our estimate for $\pi$ is therefore  
$$ \pi \approx 4.0 \frac{\# hits}{N}$$  
Let's define a function piEstimate(N) to return an estimate of $\pi$.  
I'll create arrays of x and y values randomly selected between 0 and 1  
using numpy's ranf function, then I'll create an array which represents the  
squared distance each point (x,y) is from the origin, and then we'll loop  
over these rsquared values an count how many are less than or equal to the  
circle radius 1.0. Our function returns then our estimate for $\pi$.

```python
def piEstimate(N):
    x = np.random.ranf([N])
    y = np.random.ranf([N])
    hits = 0
    rsqr = x**2 + y**2

    for value in np.nditer(rsqr):
        if value <= 1.0:
            hits = hits + 1
    estimate = 4.0*hits/N       
    return estimate
```

Now, go ahead and experiment with this function. Notice that it takes a goodly  
number of throws to estimate $\pi$ to even 3 significant figures. 

```python
%timeit piEstimate(100000)
```

Just for fun, here's another function to compute $\pi$ using essentially the same method;  
I thought it would be significantly faster than piEstimate(), and indeed it is about 100  
times faster! 

```python
def piEstimate2(N):
    hits = (np.random.ranf(N) ** 2 + np.random.ranf(N) ** 2 <= 1.0).sum()
    estimate = 4.0*hits/N
            
    return estimate
```

```python
%timeit piEstimate2(100000)
```
