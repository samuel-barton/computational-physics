```python
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
from matplotlib.ticker import LinearLocator, FormatStrFormatter

def plot3d(f, xmin,xmax, dx, ymin, ymax, dy, outfile='None'):
    x = np.arange(xmin,xmax+dx,dx)
    y = np.arange(ymin,ymax+dy,dy)
    x,y = np.meshgrid(x,y)
    z = f(x,y)
    z = f(x,y)
    fig = plt.figure(figsize=(14,10))
    ax = fig.gca(projection='3d')
    cset = ax.contourf(x, y, z, 20, zdir='z', offset=-0.4, cmap=cm.coolwarm)
    ax.plot_surface(x, y, z, rstride=1, cstride=1, cmap=cm.coolwarm,
       linewidth=0.01, antialiased=True, alpha = .85)
    ax.set_xlabel(r'$\mathrm{x-axis}$', fontsize=24)
    ax.set_xlim(-2.1,2.1)
    ax.set_ylabel(r'$\mathrm{y-axis}$', fontsize=24)
    ax.set_ylim(-2.1, 2.1)
    ax.set_zlabel(r'$\mathrm{z-axis}$', fontsize=24)
    ax.set_zlim(-0.4, 0.4)
    if outfile != 'None': plt.savefig(outfile)
    plt.show()
    
```

```python
def f(x,y):
    return 2*np.sin(x*y)*np.exp(-(x**2 + y**2))
```

```python
plot3d(f,-2,2,.1, -2,2,.1)
```
