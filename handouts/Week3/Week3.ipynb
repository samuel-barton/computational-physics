### NumPy
How to create arrays (1d)

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
array_1d = np.array([0,1.7,4.3,5.8,8.4,10])
print(type(array_1d))
print(array_1d)
print(array_1d.ndim)
plt.plot(array_1d)
plt.show()
```

You probably don't want to plot the graph above (unless you **really** believe the underlying phenomena is erratic)
It's better to plot the points and fit the data to some model. For now, let's just modify this to plot the points themselves. 

```python
% matplotlib notebook
plt.plot(array_1d, 'ro')
plt.show()
```
